# TF + Linear Regression

```python
import tensorflow as tf
import numpy as np
```

```python
# Generation of datapoints based on a linear model: y = W.x + b
actual_W = 2
actual_b = 1
datapoints_count = 1000
xs = []
ys = []
for i in range(datapoints_count):
    xs.append(i)
    ys.append(actual_W*i+actual_b)
```

```python
# Model
x = tf.placeholder(tf.float32) # features
W = tf.Variable(0.0, name="weight")
b = tf.Variable(0.0, name="bias")
y = tf.multiply(x, W) + b # output

# labels
y_ = tf.placeholder(tf.float32)

# Cost, Optimizer
cost = tf.reduce_mean(tf.square(y_-y))
learning_rate = 0.0000001
train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
tf.summary.scalar("cost_history", cost)
tf.summary.scalar("W_history", W)
tf.summary.scalar("b_history", b)
merged_summaries = tf.summary.merge_all()
```

```python
init = tf.global_variables_initializer()
```

## Stochastic Gradient Descent (SGD)

```python
writerSGD = tf.summary.FileWriter("./logs/sgd")
```

```python
sess = tf.Session()
sess.run(init)

batch_size = 1

for i in range(1000):

    print "Epoch:", i

    batch_start_idx = (i * batch_size)
    batch_end_idx = batch_start_idx + batch_size

    batch_xs = xs[batch_start_idx: batch_end_idx]
    batch_ys = xs[batch_start_idx: batch_end_idx]

    # Train
    feed = {x: batch_xs, y_: batch_ys}
    _, summary = sess.run([train_step, merged_summaries], feed_dict=feed)

    writerSGD.add_summary(summary, i)

    print("Cost: %f" % sess.run(cost, feed_dict=feed))
        
sess.close()
```

## Mini-Batch Gradient Descent (MBGD)

```python
writerMBGD = tf.summary.FileWriter("./logs/mbgd")
```

```python
sess = tf.Session()
sess.run(init)

batch_size = 5

for i in range(200):

    print "Epoch:", i

    batch_start_idx = (i * batch_size)
    batch_end_idx = batch_start_idx + batch_size

    batch_xs = xs[batch_start_idx: batch_end_idx]
    batch_ys = xs[batch_start_idx: batch_end_idx]
    # print "batch_xs:", batch_xs
    # print "batch_ys:", batch_ys

    # Train
    feed = {x: batch_xs, y_: batch_ys}
    _, summary = sess.run([train_step, merged_summaries], feed_dict= feed)

    writerMBGD.add_summary(summary, i)

    print("Cost: %f" % sess.run(cost, feed_dict=feed))
        
sess.close()
```

## Batch Gradient Descent (BGD)

```python
writerBGD = tf.summary.FileWriter("./logs/bgd")
```

```python
sess = tf.Session()
sess.run(init)

for i in range(300):

    print "Epoch:", i

    # Train
    feed = {x: xs, y_: ys}
    _, summary = sess.run([train_step, merged_summaries], feed_dict= {x: xs, y_: ys})

    writerBGD.add_summary(summary, i)

    print("Cost: %f" % sess.run(cost, feed_dict=feed))
        
sess.close()
```
