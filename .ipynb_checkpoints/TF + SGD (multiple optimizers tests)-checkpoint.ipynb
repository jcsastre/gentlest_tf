# TF + SGD (multiple optimizers tests)

```python
import tensorflow as tf
import numpy as np
```

```python
# Training function
def train(summary_file_writer):

    sess = tf.Session()
    sess.run(init)

    for i in range(len(xs)):

        print "Epoch:", i

        # Train
        feed = {x: xs[i], y_: ys[i]}
        _, summary = sess.run([train_step, merged_summaries], feed_dict=feed)

        summary_file_writer.add_summary(summary, i)

        print("Cost: %f" % sess.run(cost, feed_dict=feed))

    sess.close()
```

```python
# Generation of datapoints based on a linear model: y = W.x + b
actual_W = 2
actual_b = 1

DATAPOINTS_COUNT = 1000
xs = []
ys = []
for i in range(DATAPOINTS_COUNT):
    xs.append(i)
    ys.append(actual_W*i+actual_b)
xs = np.transpose([xs])
ys = np.transpose([ys])    

# Graph
x = tf.placeholder(tf.float32) # features (input)
W = tf.Variable(0.0, name="weight")
b = tf.Variable(0.0, name="bias")
y = tf.multiply(x, W) + b # output
y_ = tf.placeholder(tf.float32) # labels

# Cost, Optimizer
cost = tf.reduce_mean(tf.square(y_-y))
learning_rate = 0.0000001


# Summaries
tf.summary.scalar("cost_history", cost)
tf.summary.scalar("W_history", W)
tf.summary.scalar("b_history", b)
merged_summaries = tf.summary.merge_all()

init = tf.global_variables_initializer()
```

## SGD + GradientDescentOptimizer + Fixed learning rate

```python
train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

train(tf.summary.FileWriter("./logs/SgdGdoFlr"))
```

## SGD + GradientDescentOptimizer + Variable Learning Rate

```python
# from learning_rate = 0.0000001 to 0.000001
initial_learning_rate = 0.000001

# Optimizer: set up a variable that's incremented once per batch and 
# controls the learning rate decay.
global_step = tf.Variable(initial_value=0.0, trainable=False, name = 'global_step')

learning_rate = tf.train.exponential_decay(
    learning_rate = initial_learning_rate,    # Initial learning rate.
    global_step   = global_step,      # Current index into the dataset.
    decay_steps   = DATAPOINTS_COUNT,
    decay_rate    = 0.95,               
    staircase     = True,             # If True decay the learning rate at discrete intervals.
    name = 'learning_rate'
)

train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost, global_step=global_step)

# We have defined global_step, so:
init = tf.global_variables_initializer()

train(tf.summary.FileWriter("./logs/SgdGdoVlr"))
```

### SGD + AdamOptimizer

```python
train_step = tf.train.AdamOptimizer().minimize(cost)
# The AdamOptimizer class creates additional variables, called "slots", to hold values 
# for the "m" and "v" accumulators. These variables must be initialized before you can train a model.
# (Source: http://stackoverflow.com/a/33846659/4597696)
init = tf.global_variables_initializer()

train(tf.summary.FileWriter("./logs/SgdAo"))
```
