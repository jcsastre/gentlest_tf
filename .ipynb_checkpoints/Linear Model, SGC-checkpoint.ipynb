# TF + Linear Regression

```python
import tensorflow as tf
import numpy as np
```

```python
# Generation of datapoints based on a linear model: y = W.x + b
actual_W = 2
actual_b = 1
datapoints_count = 1000
xs = []
ys = []
for i in range(datapoints_count):
    xs.append(i)
    ys.append(actual_W*i+actual_b)
```

```python
# Training function
def training(xs, ys, datapoints_count_per_epoch, train_step, summaries, log_writer):
    
    datapoints_count = len(xs)
    print "Number of datapoints:", datapoints_count
    
    epochs_required = datapoints_count / datapoints_count_per_epoch
    print "Required number of epochs:", epochs_required
    
    batch_size = datapoints_count / epochs_required
    print "Batch size:", batch_size
    
    for i in range(epochs_required):
        
        print "Epoch:", i
                    
        batch_start_idx = (i * batch_size)
        batch_end_idx = batch_start_idx + batch_size
        
        batch_xs = xs[batch_start_idx: batch_end_idx]
        batch_ys = xs[batch_start_idx: batch_end_idx]
        print "batch_xs:", batch_xs
        print "batch_ys:", batch_ys
    
        # Train
        _, summary = sess.run([train_step, summaries], feed_dict= {x: batch_xs, y_: batch_ys})
    
        writerSGD.add_summary(summary, i)
```

## Stochastic Gradient Descent (SGD)

```python
# Model
x = tf.placeholder(tf.float32) # features
W = tf.Variable(0.0, name="weight")
b = tf.Variable(0.0, name="bias")
y = tf.multiply(x, W) + b # output

# labels
y_ = tf.placeholder(tf.float32)

# Cost, Optimizer
cost = tf.reduce_mean(tf.square(y_-y))
learning_rate = 0.0000001
train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
tf.summary.scalar("cost_history", cost)
tf.summary.scalar("W_history", W)
tf.summary.scalar("b_history", b)
merged_summaries = tf.summary.merge_all()
```

```python
writerSGD = tf.summary.FileWriter("./logs/sgd")
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
training(xs, ys, datapoints_count_per_epoch=1, train_step=train_step, summaries=merged_summaries, log_writer=writerSGD)
sess.close()
```
